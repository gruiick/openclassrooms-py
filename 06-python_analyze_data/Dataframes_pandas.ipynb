* Remplacer 'NaN' par '0'

```python
titanic.fillna(value={"age": 0}).age.head(10)
```

* Remplacer 'NaN' par la valeur (de la ligne) pr√©c√©dente

```python
titanic.fillna(method="pad").age.head(10)
```

* Supprimer les axes (colonnes ou lignes) qui contiennent des 'NaN'

```python
# par d√©faut, les lignes
titanic.dropna().head(10)
```

```python
# on peut demander les colonnes
titanic.dropna(axis="columns").head()
```

* Renommer une colonne

```python
titanic.rename(columns={"sex":"sexe"})
```

Ou utiliser une lambda pour appliquer √† tous les index (par exemple)

```python
f = lambda x: x+1
titanic.rename(index=f)
```

* Supprimer des axes

```python
titanic.drop(columns=["age"])
```

Toutes ces fonctions acceptent un argument 'inplace', ce qui permet de modifier directement le dataframe, sinon, une copie est renvoy√©e (peut poser des probl√®mes de m√©moire...)

* Tableaux crois√©s dynamiques (ou table pivot)

```python
titanic.pivot_table('survived', index='sex', columns='class')
# en moyenne (default)
```

```python
titanic.pivot_table('survived', index='sex', columns='class', aggfunc="sum")
# en 'unit√©s' (sum())
```

et en dataframe multi-index√© :

```python
titanic.dropna(inplace=True)
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class', aggfunc="sum")
```
