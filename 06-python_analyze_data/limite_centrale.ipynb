```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

domaine = range(-100, 100)
mu = 0
sigma = 20

f = lambda x: 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]
plot = plt.plot(domaine, y)
```

```python
# considérons 100 variables contenant chacune 200 échantillons
matrice_aleatoire = np.random.rand(100, 200)
sommes = np.sum(matrice_aleatoire, 0)
print(f'la taille de sommes est {sommes.shape}.')
```

```python
plot = plt.scatter(range(200), sommes)
```

```python
plot = plt.hist(sommes)
```

```python
# plus gros ?
matrice_aleatoire = np.random.rand(10000, 20000)
sommes = np.sum(matrice_aleatoire, 0)
plot = plt.hist(sommes, bins=100)
```

```python
print(f'La moyenne empirique de notre distribution est {np.mean(sommes)}')
print(f'La moyenne empirique de la fonction rand est {np.mean(np.random.rand(1000000))}')
print(f'La variance empirique de notre distribution est {np.var(sommes)}')
print(f'La variance empirique de la fonction rand est {np.var(np.random.rand(1000000))}')
```
